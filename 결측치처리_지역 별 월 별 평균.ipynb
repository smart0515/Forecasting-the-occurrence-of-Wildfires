```python
data=pd.read_csv("산불_datawindow_latest.csv",encoding='cp949')
```

```python
data['날짜']=pd.to_datetime(data['날짜'])
data[['계절']]=np.nan
```

```python
for i in range(2001, 2023):
    data.loc[(f"{i}-02-01" <= data['날짜']) & (data['날짜'] <= f"{i}-05-15"), '계절']=data.loc[(f"{i}-02-01" <= data['날짜']) & (data['날짜'] <= f"{i}-05-15"), '계절'].fillna(2)
    data.loc[(f"{i}-11-01" <= data['날짜']) & (data['날짜'] <= f"{i}-12-15"), '계절']=data.loc[(f"{i}-11-01" <= data['날짜']) & (data['날짜'] <= f"{i}-12-15"), '계절'].fillna(1)
data[['계절']]=data[['계절']].fillna(0)
```

```python
import datetime
data['month'] = data['날짜'].dt.strftime('%m')
```

**발생지역별 월 평균을 구해서 결측값을 채우기위해 groupby 사용**

```python
data.groupby(['발생지역','month']).mean()
data.groupby(['발생지역','month']).mean()[['상대습도']]
```

```python
cols=data.columns.drop(['날짜','발생지역','주소','경도','위도','피해면적','지점_asos', '거리_asos','지점_aws','거리_aws', '계절','month'])
cols
```

```python
for i in tqdm(range(len(cols))):
    for j in range(len(data)):
        if(np.isnan(data[[cols[i]]].loc[j].values[0])):  # 해당 컬럼의 값이 결측값이면 
            loc=data.loc[j]['발생지역']
            month=data.loc[j]['month']
            tmp=data.groupby(['발생지역','month']).mean()[['상대습도']].loc[loc].loc[month].values[0]
            print(loc,month)
            print(tmp)

            data[cols[i]].loc[j]=tmp

            print(data[[cols[i]]].loc[j])

```
