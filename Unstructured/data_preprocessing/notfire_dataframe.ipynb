### 그 지점의 인덱스 추출

```python
import os
import pandas as pd
import numpy as np

# 대상 폴더 경로 설정
folder_path = "no_fire_loc/"

# 결과를 저장할 리스트 초기화
list_i = []
list_j = []

# CSV 파일에서 값이 9999인 인덱스 찾기
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        filepath = os.path.join(folder_path, filename)
        df = pd.read_csv(filepath)
        df_9999 = df == 9999
        i_indices, j_indices = np.where(df_9999)
        list_i.extend(i_indices)
        list_j.extend(j_indices)

# 결과 출력
print(list_i)
print(list_j)
```

```python
data['j']=list_j
```

## 데이터프레임 크기에 맞게 구해둔 엑셀 경위도와 인덱스를 맵핑
- 산불난 지점의 경위도를 찾을 수 있음

```python
data_match=pd.read_csv('액셀경위도.csv')
```

```python
data_match = data_match.rename(columns={'x': 'j', 'y': 'i'})
```

```python
data_match.drop(['value'],axis=1,inplace=True)
```

```python
data_match
```

```python
lon_list = []

for i_value, j_value in zip(list_i, list_j):
    try:
        result = data_match.loc[(data_match['i'] == i_value) & (data_match['j'] == j_value), 'lon'].values[0]
    except IndexError:
        result = None
        
    lon_list.append(result)
```

```python
lat_list = []

for i_value, j_value in zip(list_i, list_j):
    try:
        result = data_match.loc[(data_match['i'] == i_value) & (data_match['j'] == j_value), 'lat'].values[0]
    except IndexError:
        result = None
        
    lat_list.append(result)
```

```python
data['lon']=lon_list
```

```python
data['lat']=lat_list
```
